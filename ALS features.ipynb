# Static features 

```python
features = {}
def readStaticData():
    with open('Dados/static_data.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count = 1
                continue
            patientRef = row[0]
            if row[3] != '' and row[25] != '':
                if patientRef in features:
                    features[patientRef].append(float(row[3])) # age
                    features[patientRef].append(float(row[25])) # family history
                else:
                    features[patientRef] = [float(row[3])]
                    features[patientRef].append(float(row[25]))
        print()
        
readStaticData()
for id in features:
    print(id)
    print(features[id])
    print()
```

```python
def drawAgeDist():
    ageList = []
    for v in features.values():
        ageList.append(v[0])
        
    print('Min: ', min(ageList))
    print('Max: ', max(ageList))
    ageAverage = round(sum(ageList) / len(ageList), 1)
    print('Average: ', ageAverage)
    deviation = np.std(ageList)
    print('Standard deviation: ', round(deviation,1))
    
    plt.hist(ageList, bins=20)
    plt.title('Distribution of patient ages')
    plt.xlabel('Ages')
    plt.ylabel('Count')
    plt.show()
    
drawAgeDist()
```
