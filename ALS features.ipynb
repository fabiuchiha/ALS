# Static features 

```python
staticFeatures = {}
def readStaticData():
    with open('Dados/static_data.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count = 1
                continue
            patientRef = row[0] # patient ref
            if row[0] != '':
                staticFeatures[patientRef] = [row[1]]
                for index in range(2,44):
                    staticFeatures[patientRef].append(row[index])
        print()
        
readStaticData()
print(len(staticFeatures))
for id in staticFeatures:
    print(id)
    print(staticFeatures[id])
    print()
```

```python
def drawNumericDist(featureIndex, title, xLabel):
    featureList = []
    for v in staticFeatures.values():
        feature = v[featureIndex]
        if feature != '':
            featureList.append(float(feature))
        
    print('Data points: ', len(featureList))
    print('Min: ', min(featureList))
    print('Max: ', max(featureList))
    featureAverage = round(sum(featureList) / len(featureList), 1)
    print('Average: ', featureAverage)
    deviation = np.std(featureList)
    print('Standard deviation: ', round(deviation,1))
    
    plt.hist(featureList, bins=20)
    plt.title(title)
    plt.xlabel(xLabel)
    plt.ylabel('Count')
    plt.show()
```

```python
drawNumericDist(1, 'Distribution of patient birth years', 'Year')
drawNumericDist(2, 'Distribution of patient ages', 'Age')
drawNumericDist(20, 'Distribution of patient heights', 'Height')
drawNumericDist(21, 'Distribution of patient weights before 1st symptoms', 'Weight')
drawNumericDist(22, 'Distribution of patient weights at 1st visit', 'Weight')
```

```python
def drawCategoricalDist(featureIndex, title, xLabel, columnLabels, dataValues):
    featureList = []
    for v in staticFeatures.values():
        feature = v[featureIndex]
        if feature != '':
            featureList.append(float(feature))
        
    print('Data points: ', len(featureList))
    countList = []
    for v in dataValues:
        count = featureList.count(v)
        countList.append(count)
        print(count)
    
    plt.bar(columnLabels, countList, color ='maroon')
    plt.title(title)
    plt.xlabel(xLabel)
    plt.ylabel('Count')
    plt.show()
```

# Functional scores

```python
frsFeatures = {}
revisedFRSFeatures = {}
def readFRSData():
    with open('Dados/ALSFRS.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count = 1
                continue
            patientRef = row[0] # patient ref
            if row[0] != '':
                frsPosition = 2
                while (row[frsPosition].isnumeric()):
                    alsfrs = row[frsPosition]
                    if patientRef in frsFeatures:
                        frsFeatures[patientRef].append(alsfrs)
                    else:
                        frsFeatures[patientRef] = [alsfrs]
                    frsPosition = frsPosition + 18
                    if frsPosition >= len(row):
                        break
                frsPosition = 3
                while (row[frsPosition].isnumeric()):
                    alsfrs = row[frsPosition]
                    if patientRef in revisedFRSFeatures:
                        revisedFRSFeatures[patientRef].append(alsfrs)
                    else:
                        revisedFRSFeatures[patientRef] = [alsfrs]
                    frsPosition = frsPosition + 18
                    if frsPosition >= len(row):
                        break
        print()
        
readFRSData()
print(len(frsFeatures))
for id in frsFeatures:
    print(id)
    print(frsFeatures[id])
    print()
```

```python
def drawTemporalDist(featureDict, featureIndex, title, xLabel):
    featureList = []
    for v in featureDict.values():
        if featureIndex >= len(v):
            continue
        feature = v[featureIndex]
        if feature != '':
            featureList.append(float(feature))
        
    print('Data points: ', len(featureList))
    print('Min: ', min(featureList))
    print('Max: ', max(featureList))
    featureAverage = round(sum(featureList) / len(featureList), 1)
    print('Average: ', featureAverage)
    deviation = np.std(featureList)
    print('Standard deviation: ', round(deviation,1))
    
    plt.hist(featureList, bins=20)
    plt.title(title)
    plt.xlabel(xLabel)
    plt.ylabel('Count')
    plt.show()
```

```python
drawTemporalDist(frsFeatures, 0, 'Distribution of ALSFRS t=0', 'ALSFRS')
drawTemporalDist(frsFeatures, 1, 'Distribution of ALSFRS t=1', 'ALSFRS')
drawTemporalDist(frsFeatures, 2, 'Distribution of ALSFRS t=2', 'ALSFRS')
drawTemporalDist(frsFeatures, 3, 'Distribution of ALSFRS t=3', 'ALSFRS')
drawTemporalDist(frsFeatures, 4, 'Distribution of ALSFRS t=4', 'ALSFRS')
```

```python
drawTemporalDist(revisedFRSFeatures, 0, 'Distribution of ALSFRS-r t=0', 'ALSFRS')
drawTemporalDist(revisedFRSFeatures, 1, 'Distribution of ALSFRS-r t=1', 'ALSFRS')
drawTemporalDist(revisedFRSFeatures, 2, 'Distribution of ALSFRS-r t=2', 'ALSFRS')
drawTemporalDist(revisedFRSFeatures, 3, 'Distribution of ALSFRS-r t=3', 'ALSFRS')
drawTemporalDist(revisedFRSFeatures, 4, 'Distribution of ALSFRS-r t=4', 'ALSFRS')
```

# Respiratory features

```python
fvcFeatures = {}
bmiFeatures = {}
def readRespiratoryData():
    with open('Dados/fvc_resp.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count = 1
                continue
            patientRef = row[0] # patient ref
            if row[0] != '':
                fvcPosition = 2
                while (row[fvcPosition] != ''):
                    fvc = row[fvcPosition]
                    if patientRef in fvcFeatures:
                        fvcFeatures[patientRef].append(fvc)
                    else:
                        fvcFeatures[patientRef] = [fvc]
                    fvcPosition = fvcPosition + 3
                    if fvcPosition >= len(row):
                        break
                bmiPosition = 3
                while (row[bmiPosition] != '' and isfloat(row[bmiPosition])):
                    bmi = row[bmiPosition]
                    if patientRef in bmiFeatures:
                        bmiFeatures[patientRef].append(bmi)
                    else:
                        bmiFeatures[patientRef] = [bmi]
                    bmiPosition = bmiPosition + 3
                    if bmiPosition >= len(row):
                        break
        print()
        
readRespiratoryData()
print(len(bmiFeatures))
for id in bmiFeatures:
    print(id)
    print(bmiFeatures[id])
    print()
```

```python
drawTemporalDist(fvcFeatures, 0, 'Distribution of FVC t=0', 'FVC')
drawTemporalDist(fvcFeatures, 1, 'Distribution of FVC t=1', 'FVC')
drawTemporalDist(fvcFeatures, 2, 'Distribution of FVC t=2', 'FVC')
drawTemporalDist(fvcFeatures, 3, 'Distribution of FVC t=3', 'FVC')
drawTemporalDist(fvcFeatures, 4, 'Distribution of FVC t=4', 'FVC')
```

```python
drawTemporalDist(bmiFeatures, 0, 'Distribution of BMI t=0', 'BMI')
drawTemporalDist(bmiFeatures, 1, 'Distribution of BMI t=1', 'BMI')
drawTemporalDist(bmiFeatures, 2, 'Distribution of BMI t=2', 'BMI')
drawTemporalDist(bmiFeatures, 3, 'Distribution of BMI t=3', 'BMI')
drawTemporalDist(bmiFeatures, 4, 'Distribution of BMI t=4', 'BMI')
```
