```python
snapshots = {}
def readSnapshotsData():
    with open('Dados/snapshots_independent.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count = 1
                continue
            patientRef = row[0] # patient ref
            snapshotList = []
            for index in range(1,61):
                snapshotList.append(row[index])
            if patientRef in snapshots:
                snapshots[patientRef].append(snapshotList)
            else:
                snapshots[patientRef] = [snapshotList]
            
        print()
        
readSnapshotsData()
print('Number of patients with snapshots:', len(snapshots))
print('Percentage:', round(len(snapshots)/1632*100, 1))
#for id in snapshots:
 #   print(id)
  #  print(snapshots[id])
   # print()
```

```python
def drawTemporalDist(snapshotIndex, featureIndex, title, xLabel):
    featureList = []
    for v in snapshots.values():
        if snapshotIndex >= len(v):
            continue
        feature = v[snapshotIndex][featureIndex]
        if feature != '':
            featureList.append(float(feature))
        
    print('Data points:', len(featureList))
    print('Percentage:', round(len(featureList)/1560*100, 1))
    print('Min:', min(featureList))
    print('Max:', max(featureList))
    featureAverage = round(sum(featureList) / len(featureList), 1)
    print('Average:', featureAverage)
    deviation = np.std(featureList)
    print('Standard deviation:', round(deviation,1))
    
    plt.hist(featureList, bins=20)
    plt.title(title)
    plt.xlabel(xLabel)
    plt.ylabel('Count')
    plt.show()
```

```python
drawTemporalDist(0, 58, 'Distribution of ALSFRS-r t=0', 'ALSFRS-r') # ALSFRS-r
drawTemporalDist(1, 58, 'Distribution of ALSFRS-r t=1', 'ALSFRS-r')
drawTemporalDist(2, 58, 'Distribution of ALSFRS-r t=2', 'ALSFRS-r')
drawTemporalDist(3, 58, 'Distribution of ALSFRS-r t=3', 'ALSFRS-r')
```
